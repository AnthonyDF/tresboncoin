```python
import pandas as pd
```

```python
df = pd.read_csv('March_2021_MPSOV.csv')
df.head()
```

```python
print('Manufacturer count :',df.Make.nunique())
print('Model count :',df.Model.nunique())
print('Submodel count :',df.Submodel.nunique())
```

```python
df.Make.value_counts()
```

```python
def normalize(df, column):
    
    df[f'{column}_TBC'] = df[f'{column}'].str.lower().replace(' ','')
    df[f'{column}_TBC'] = df[f'{column}_TBC'].str.strip()
    df[f'{column}_TBC'] = df[f'{column}_TBC'].str.replace('\s','')
    df[f'{column}_TBC'] = df[f'{column}_TBC'].str.replace('-','')
    df[f'{column}_TBC'] = df[f'{column}_TBC'].str.replace('_','')
    
    df[f'{column}_TBC_inv'] = df[f'{column}'].apply(lambda x: "".join(list(reversed(x.split()))))
    df[f'{column}_TBC_inv'] = df[f'{column}_TBC_inv'].str.lower()
    df[f'{column}_TBC_inv'] = df[f'{column}_TBC_inv'].str.replace('\s','')
    df[f'{column}_TBC_inv'] = df[f'{column}_TBC_inv'].str.replace('-','')
    df[f'{column}_TBC_inv'] = df[f'{column}_TBC_inv'].str.replace('_','')
    df[f'{column}_TBC_inv'] = df[f'{column}_TBC_inv'].str.replace(' ','')
    
    return df
```

```python
columns_to_implement = ['Make','Model','Model_Submodel','Submodel']

for col in columns_to_implement:
    df = normalize(df, col)
df.head()
```

```python
df.to_csv('March_2021_MPSOV.csv', index=False)
```
