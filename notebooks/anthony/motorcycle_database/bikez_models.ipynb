```python
from datetime import datetime
import requests
from bs4 import BeautifulSoup
import time
import random
import pandas as pd
import os
import random
import numpy as np
```

# SCRAP DATA

```python
# site to scrap
source = 'bikez'

# url to scrap
url = 'https://bikez.com/brands/index.php'
response = requests.get(url)
    
soup = BeautifulSoup(response.content, "html.parser")

brand_count = 0

for brand_ in soup.find('table', class_="zebra").find_all('td')[2:]:
    try:
        url_brand = "https://bikez.com" + brand_.find("a").get('href')[2:]
        brand = brand_.find("a").text.replace(' motorcycles', '')
        print('BRAND:', brand.upper())
        
        response_brand = requests.get(url_brand)
        soup_brand = BeautifulSoup(response_brand.content, "html.parser")
        
        for model_ in soup_brand.find('table', class_="zebra").find_all('td'):
            for test in model_.find_all('a'):
                model = test.text
                if len(model) == 0:
                    pass
                else:
                    model = model.replace(brand + ' ','')
                    print(model)
                    features_dict = {
                            'brand':[brand], 
                            'model':[model]}

                    df = pd.DataFrame(features_dict)
                    df.to_csv(f"model/{features_dict['brand'][0]}-{features_dict['model'][0]}.csv")
        print('-----------')
                    
    except:
        pass
```

# BUILD DATAFRAME AND CLEAN

```python
history = pd.DataFrame(
    data = {'brand':[], 
            'model':[]})

for filename in [file for file in os.listdir('model') if file.endswith(".csv")]:
    df = pd.read_csv(f"model/{filename}")
    history = history.append(df)

history.to_csv("bikez_models.csv", index=False)
```
