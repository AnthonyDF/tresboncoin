```python
from datetime import datetime
import requests
from bs4 import BeautifulSoup
import time
import random
import pandas as pd
import os
import random
import numpy as np
```

# SCRAP DATA

```python
# site to scrap
source = 'motorcycledb'

# url to scrap
url = 'https://www.motorcycledb.com/manufacturers.php'
response = requests.get(url)
    
soup = BeautifulSoup(response.content, "html.parser")

brand_soup = soup.find_all('div', class_="makes")
count_brand = 0

for brand in brand_soup [36+144:]:
    brand_url= 'https://www.motorcycledb.com' + brand.find("a").get("href")
    brand_name =  brand_url.split("/")[-2]
    response_brand = requests.get(brand_url)
    brand_url_soup = BeautifulSoup(response_brand.content, "html.parser")
    
    print('brand:', count_brand)
    print(brand_url)
    
    count_model = 0
    
    model_soup = brand_url_soup.find('div', class_="news_item_title").find_all('div', class_="row")
    for row in model_soup:
        for model in row.find_all('a'):
            try:
                model_url= 'https://www.motorcycledb.com' + model.get("href")
                print('model', model_url)
                print('model',count_model)

                response_model = requests.get(model_url)
                model_url_soup = BeautifulSoup(response_model.content, "html.parser")
                model_table = model_url_soup.find('table')
                table_rows  = model_table.find_all("td")

                data = list()

                for row in table_rows:
                    data.append(row.text)

                features = ['Power:', 
                            'Category:', 
                            'Year:', 
                            'Engine type:', 
                            'Model:', 
                            'Overall width:', 
                            'Overall length:',
                            'Dry weight:',
                            'Power/weight ratio:',
                            'Fuel capacity:',
                            'Fuel system:',
                            'Capacity:']

                features_dict = {
                    'Brand':[brand_name],
                    'Power:':[np.nan], 
                    'Category:':[np.nan], 
                    'Year:':[np.nan], 
                    'Engine type:':[np.nan], 
                    'Model:':[np.nan], 
                    'Overall width:':[np.nan], 
                    'Overall length:':[np.nan],
                    'Dry weight:':[np.nan],
                    'Power/weight ratio:':[np.nan],
                    'Fuel capacity:':[np.nan],
                    'Fuel system:':[np.nan],
                    'Capacity:':[np.nan]
                    }

                for feature in features:
                    try:
                        position = data.index(feature)+1
                        features_dict[feature]= [data[position]]
                    except:
                        pass
                df = pd.DataFrame(features_dict)
                df.to_csv(f"bike/{features_dict['Model:']}-{features_dict['Year:']}.csv")
            except:
                pass
            
            count_model += 1
    count_brand += 1
```

# BUILD DATAFRAME AND CLEAN

```python
history = pd.DataFrame(
    data = {
            'Brand':[],
            'Power:':[], 
            'Category:':[], 
            'Year:':[], 
            'Engine type:':[], 
            'Model:':[], 
            'Overall width:':[], 
            'Overall length:':[],
            'Dry weight:':[],
            'Power/weight ratio:':[],
            'Fuel capacity:':[],
            'Fuel system:':[],
            'Capacity:':[]
            })

for filename in [file for file in os.listdir('bike') if file.endswith(".csv")]:
    df = pd.read_csv(f"bike/{filename}")
    history = history.append(df)

history.to_csv("master_model_motorcycledb.csv", index=False)
```

```python
df = pd.read_csv('master_model_motorcycledb.csv')
new_columns_name = [column.lower().replace(':','') for column in df.columns]
df.columns=new_columns_name
df.dropna(subset=['model'], inplace=True)
# brand and model cleaning
df['raw_model'] = df.model
df['raw_brand'] = df.brand
df.model = df.model.str.lower()
df.brand = df.brand.str.lower()
df.model = df.model.str.strip()
df.brand = df.brand.str.strip()
df['model'] = df.apply(lambda x: x['model'].replace(x['brand'],''), axis=1)
df.model = df.model.str.strip()
df.model = df.model.str.replace(r"\(.*\)","").str.replace('[^\w\s]','').str.replace(" ","")
df.brand = df.brand.str.replace(r"\(.*\)","").str.replace('[^\w\s]','').str.replace(" ","")
df['brand_model'] = df.brand + df.model + df.year.astype(str)
df.brand_model = df.brand_model.str.replace('.0', '', regex=False)

#clean category
df.category = df.category.str.lower().str.strip().str.split(',')

#clean engine type
df['engine type'] = df['engine type'].str.lower().str.strip().str.split(',')

#clean width length and weight, power ratio, fuel capacity
def parser(to_clean, delimiter):
    try:
        if type(to_clean)==float:
            return np.nan
        cleaned = to_clean.split(f' {delimiter}')[0].replace(",","")
        return float(cleaned)
    except:
        return np.nan

df['width_mm'] = df.apply(lambda x: parser(x['overall width'],'mm'), axis=1)
df['length_mm'] = df.apply(lambda x: parser(x['overall length'],'mm'), axis=1)
df['weight_kg'] = df.apply(lambda x: parser(x['dry weight'],'kg'), axis=1)
df['power/weight ratio'] = df.apply(lambda x: parser(x['power/weight ratio'],'HP/kg'), axis=1)
df['fuel_capacity_ltr'] = df.apply(lambda x: parser(x['fuel capacity'],'litres'), axis=1)
df['capacity_ccm'] = df.apply(lambda x: parser(x['capacity'],'ccm'), axis=1)
df['power_hp'] = df.apply(lambda x: parser(x['power'],'Horsepower'), axis=1)

# engine type
def parser_engine_type(engine_type):
    if type(engine_type)==float:
        return np.nan
    return engine_type[0]

df['engine_type'] = df.apply(lambda x: parser_engine_type(x['engine type']), axis=1)

df.drop(columns=['power', 'engine type',
                 'overall width', 'overall length', 
                 'dry weight','fuel capacity', 'capacity'], 
        inplace=True)

df.columns = [x+"_db" for x in df.columns]

df.dropna(subset=['year_db'],inplace=True)
df.year_db = df.year_db.astype(int)
```
